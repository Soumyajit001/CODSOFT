## Initialize Text Vectorization(TF-IDF) vectorizer

```python
# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Fit and transform the training data
x_train_tfidf = tfidf_vectorizer.fit_transform(df_train['Description'])

# Transform the test data
y_train = df_train['Genre']
```

## Split Data and Train Model

```python
x = x_train_tfidf
y = df_train['Genre']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print("\nsize of x: ", x.shape)
print("\nsize of y: ", y.shape)
```

```python
print("\nsize of x_train: ", x_train.shape)
print("\nsize of x_test: ",x_test.shape)
print("\nsize of y_train: ",y_train.shape)
print("\nsize of x_test: ",y_test.shape)
```

```python
model_1 = MultinomialNB()
model_1.fit(x_train, y_train)
```

```python
y_train_prediction1 = model_1.predict(x_train)

print("Accuracy on training set: ", accuracy_score(y_train, y_train_prediction1))

print("\nClassification Report on training set:\n", classification_report(y_train, y_train_prediction1))
```

```python
model_2 = BernoulliNB()
model_2.fit(x_train, y_train)
```

```python
y_train_prediction2 = model_2.predict(x_train)

print("Accuracy on training set: ", accuracy_score(y_train, y_train_prediction2))

print("\nClassification Report on training set:\n", classification_report(y_train, y_train_prediction2))
```

```python
model_3 = MLPClassifier()
model_3.fit(x_train, y_train)
```

```python
y_train_prediction3 = model_3.predict(x_train)

print("Accuracy on training set: ", accuracy_score(y_train, y_train_prediction3))

print("\nClassification Report on training set:\n", classification_report(y_train, y_train_prediction3))
```

```python
columns = ['Multinomial NB', 'Bernoulli NB', 'MLP Classifier']
accuracies=[]
```

```python
def call(NB_model):
    NB_model.fit(x_train, y_train)
    NB_prediction = NB_model.predict(x_test)
    NB_accuracy = accuracy_score(y_test, NB_prediction)

    accuracies.append(NB_accuracy)
    
    print("________MODEL EVALUATION ON TEST DATA________")
    print("Model: ")
    print(NB_model)
    print()
    print("Accuracy of the Model: ")
    print(NB_accuracy)
```
